```python
import pandas as pd
import os

reg_file = '../visuals/regression_comparison.csv'
clf_file = '../visuals/classification_comparison.csv'

print("Loading Comparison Results...")
```

```python
# Best regression model
if os.path.exists(reg_file):
    reg_df = pd.read_csv(reg_file)
    # Sort by MAE (Ascending) -> Lowest Error is best
    best_reg = reg_df.sort_values(by='MAE', ascending=True).iloc[0]
    
    
    print(f"\nBEST REGRESSION MODEL: {best_reg['Algorithm']}")
    print(f"MAE: {best_reg['MAE']}")
    print(f"R2 Score: {best_reg['R2']}")
    print("Reasoning: Lowest Mean Absolute Error on Test data.\n")
    
else:
    print("Regression results not found. Please run 02_regression.ipynb.")
```

```python
# Best classification model
if os.path.exists(clf_file):
    clf_df = pd.read_csv(clf_file)
    # Sort by F1 Score (Descending) -> Highest is best
    best_clf = clf_df.sort_values(by='F1 Score', ascending=False).iloc[0]
    
   
    print(f"BEST CLASSIFICATION MODEL: {best_clf['Algorithm']}")
    print(f"F1 Score: {best_clf['F1 Score']}")
    print(f"Accuracy: {best_clf['Accuracy']}")
    print("Reasoning: Highest F1 Score (balanced precision/recall).")
    
else:
    print("Classification results not found. Please run 03_classification.ipynb.")
```

```python
report_content = """AutoML Analysis Summary"""
if os.path.exists(reg_file):
    report_content += f"\nBest Regression: {best_reg['Algorithm']} (MAE: {best_reg['MAE']})\n"
if os.path.exists(clf_file):
    report_content += f"Best Classification: {best_clf['Algorithm']} (F1: {best_clf['F1 Score']})\n"

print(report_content)

```
